```python
import pandas as pd
df = pd.read_csv('DatosSeguros.csv')
df.info()
```

```python
df.isnull().sum()
```

#vamos a eliminar los dantos nulos, pues estos representan una porcion peque√±a de todos lo datos


```python
df=df.dropna()

df.info(),

#Ahi ya tienen todas las variables el mismo numero de registros
```

```python
#ESTO ES PARA ELIMINAR VALORES DUPLICADOS
df=df.drop_duplicates()
df.info()
```

```python
#Datos por edad de menor a mayor
edad= df.sort_values('edad')
edad.head(10)
```

```python
# Ordenar datos de mayor a menor por la columna 'edad'
edad = edad.sort_values('edad', ascending=False)

# Mostrar las primeras 10 filas
edad.head(10)
```

```python
# Definir los rangos de edades
rangos = [17, 28, 38, 48, 58, 68]

# Establecer el nombre para cada rango
nombrerango = ["A", "B", "C", "D", "E"]

# Crear nueva variable con los rangos de edad
df['Rango_Edad'] = pd.cut(df['edad'], rangos, labels=nombrerango)

df.head()
```

```python
df.hist(bins=20,figsize=(20,10))
```

```python
df.hist(bins=20,figsize=(20,10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df.imc,color="b",bins=30,kde=True)
plt.show()
```

```python
plt. Figure(figsize=(10,7))
sns.countplot(df.Rango_Edad)
plt.show()
```

```python
#Distribuci√≥n porcentual de hijos
total_hijos = df['hijos'].groupby(df['hijos']).count()
etiquetas = [0,1,2,3,4,5]
colors = sns.color_palette('pastel')[0:6]
plt.pie(total_hijos, labels = etiquetas, colors = colors,
autopct='%.0f%%')
plt.show()
```

```python
#Distribuci√≥n porcentual rango de edad
total_rango_edad = df['Rango_Edad'].groupby(df['Rango_Edad']).count()
labels = ["A", "B", "C", "D", "E"]
colors = sns.color_palette('Set2')[0:5]
plt.pie(total_rango_edad, labels = labels, colors = colors, autopct='%.1f%%')
plt.show()
```

```python
#establecer la relaci√≥n entre el valor del seguro y el rango de edades
sns.boxplot(x=df["Rango_Edad"],
y=df["valor_seguro"])
```

```python
#establecer la relaci√≥n entre el valor del seguro y el estado fumador 
fumador_valor = sns.boxplot(x=df["fumador"],
y=df["valor_seguro"])
```

```python
#Hay muchos outliers (c√≠rculos encima del bigote), lo que significa que hay varias personas no fumadoras con seguros m√°s caros de lo esperado, 
#aunque siguen siendo menores que en fumadores.

#S√≠ hay relaci√≥n clara: ser fumador est√° asociado con pagar primas de seguro mucho m√°s altas.

#Variabilidad mayor en fumadores: no solo pagan m√°s en promedio, sino que los valores entre fumadores var√≠an mucho m√°s que en no fumadores.

total_rango_edad = df['Rango_Edad'].groupby(df['Rango_Edad']).count()
labels = ["A", "B", "C", "D", "E"]
colors = sns.color_palette('Set2')[0:5]

plt.pie(total_rango_edad, labels = labels, colors = colors, autopct='%.1f%%')
plt.show()
```

```python

# Calcular la correlaci√≥n solo con variables num√©ricas
correlacion = df.corr(numeric_only=True)

# Mostrar la matriz de correlaci√≥n como mapa de calor
plt.figure(figsize=(8,6))
sns.heatmap(correlacion, xticklabels=correlacion.columns, yticklabels=correlacion.columns, annot=True, cmap="coolwarm")
plt.show()

#la relaci√≥n m√°s cercana al valor n√∫mero 1 es la edad y el √≠ndice de masa corporal con respecto al valor pagado al seguro; 
#estas relaciones toman m√°s importancia en la medida en que el valor sea m√°s cercano al n√∫mero 1.
```

```python
#Relaciones multivariado: para realizar el an√°lisis multivariado que implica m√°s de dos variables

sns.relplot(x='Rango_Edad' ,y='valor_seguro',hue='imc',data=df)

#Relaci√≥n tres variables: valor seguro, rango de edad y el √≠ndice de masa corporal

# los valores m√°s altos pagados corresponden a todos los rangos de edad con un √≠ndice de masa corporal alto.
```

```python
#Relaciona tres variables: valor seguro, rango de edad y el indicador de fumador

sns.relplot(x='Rango_Edad',
y='valor_seguro',hue='fumador',data=df)

#en cualquier rango de edad, el hecho de que sea fumador es determinante para el pago del valor del seguro.
```

```python
#Relaci√≥n rango de edad y estado fumador con los valores del seguro
sns.barplot(data=df,x='region',y='valor_seguro',hue='fumador')
```
